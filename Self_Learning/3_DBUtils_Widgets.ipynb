```python
dbutils.widgets.help()

```

```python
dbutils.widgets.text("country",'')
```

```python
country_name = dbutils.widgets.get("country")
print("country_name = ", country_name)
print(f"fetching {country_name} data from database") #actually here we aren't reading from DB, just msg
```

```python
dbutils.widgets.text("range_value", "5")
```

```python
range_int = int(dbutils.widgets.get("range_value"))
print(f"range = {range_int}")
for i in range(range_int):
  #print(f"i = {i}")
  print(i)
```

```python
dbutils.widgets.dropdown("Gender", "Male", ["Male", "Female"])
```

```python
gender = dbutils.widgets.get("Gender")
print(f"gender = {gender}") 
```

```python
dbutils.widgets.combobox("Color", "Red", ["Red", "Green", "Blue"])

color = dbutils.widgets.get("Color")
print(f"selected color = {color}")
```
