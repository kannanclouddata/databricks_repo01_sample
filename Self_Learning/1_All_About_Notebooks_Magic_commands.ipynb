## Important Magic cmds - ## Header2
- %md - Allows to write to write markdown text and design your notebook
- %python/ %scala - switch to Python lang in the code cell 
- %scala - but scala currently not supported in Serverless envi
- %sql - switch code cell to sql cell from any lang/others
- %sh - allows to run linux/shell cmds
- %fs - allows to interact with DBFS (DataBricks File System)
- %pip - install Python packages
- %run - call another Notebook/Python
