# Load Data into Table  -> Lakhouse formation

## cloud storage (s3)   + databricks sql  ( delta )  -> lakehouse 

```python
ddl_str="custid int, fname string, lname string, age int, profession string"
cust_df=spark.read.schema(ddl_str).csv("/Volumes/izwd37dev/wd37db/rawdatta/custs")

# writing into the table 
cust_df.write.saveAsTable("izwd37dev.wd37db.tbl_customer")
```

write dataframe into serilized file format (parquet )
-- format to byte  format 

### parquet 

**columnar** serilized , intelligent , self describing file format 

```python
ddl_str="custid int, fname string, lname string, age int, profession string"
cust_df=spark.read.schema(ddl_str).csv("/Volumes/izwd37dev/wd37db/rawdatta/custs")
#display(cust_df.take(5))
display(cust_df.take(5))

# write to parquet 

cust_df.write.mode("overwrite").parquet("/Volumes/izwd37dev/wd37db/rawdatta/cust_parquet_out")
# part-00000-tid-7560529016580187698-63729798-a7a0-4f3b-a945-a05958b3b737-144-1.c000.snappy.parquet
```

```python
%fs head /Volumes/izwd37dev/wd37db/rawdatta/cust_parquet_out/part-00000-tid-7560529016580187698-63729798-a7a0-4f3b-a945-a05958b3b737-144-1.c000.snappy.parquet
```
